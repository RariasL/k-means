```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
datos = np.random.rand(100,2)
```

```python
datos
```

```python
kmean = KMeans(n_clusters=3,random_state=42)
kmean.fit(datos)
```

```python
centro = kmean.cluster_centers_
```

```python
centro
```

```python
etiqueta = kmean.labels_
```

```python
etiqueta
```

```python
plt.figure(figsize=(10,6))
plt.scatter(datos[:,0],datos[:,1],c=etiqueta,s=50, cmap='viridis')
plt.scatter(centro[:,0],centro[:,1],marker='+',s=200,c='black')
plt.title('KMeans')
plt.show()
```
